### **Bài toán 1:**
- Một cửa hàng trong một năm (365 ngày) bán mỗi ngày từ 10 đến 40 sản phẩm
---


```python
## Thư viện
import numpy             as np
import matplotlib.pyplot as plt

from scipy.stats         import stats, uniform
```

```python
## Mô hình hóa: các giá trị tham số
n = 365 # trials
a = 10  # Min
b = 40  # Max
```

```python
## Giả lập số liệu bán hàng của 365 ngày (tạo số liệu ngẫu nhiên)
data_uniform = np.random.uniform(a, b + 1, n).astype(int)

## from scipy.stats import uniform
## data_uniform = uniform.rvs(size = n, loc = a, scale = b - a).astype(int)

data_uniform
```

```python
## Vẽ biểu đồ histogram
plt.hist(data_uniform)
plt.title('Histogram (Uniform Distribution)')
plt.show()
```

```python
## Tính xác suất 1 ngày bán không quá 15 sản phẩm: P(X <= 15) = F(15) = CDF(15)
p = (15 - a) / (b - a)
p
```

```python
## Lưu ý: Hàm scipy.stats.uniform() áp dụng cho biến LIÊN TỤC !
dist = uniform(a, b - a) # (loc = a, loc + scale = b - a)
p    = dist.cdf(15)
p
```

```python
## Tính xác suất 1 ngày bán từ 15 đến 30 sản phẩm: P(15 <= X <= 30) = F(30) - F(15)
p = (30 - 15) / (b - a)
p
```

```python
print(dist.cdf(30) - dist.cdf(15))
```

```python
## Tính xác suất 1 ngày bán nhiều hơn 20 sản phẩm: P(X > 20) = 1 - P(X <= 20) = 1 - F(20)
p = (20 - a) / (b - a)
1 - p

##
```
